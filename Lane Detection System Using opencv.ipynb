# For Image 1)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def color_selection(image):
    # Assuming white lanes
    lower_white = np.array([200, 200, 200], dtype=np.uint8)
    upper_white = np.array([255, 255, 255], dtype=np.uint8)
    mask = cv2.inRange(image, lower_white, upper_white)
    return cv2.bitwise_and(image, image, mask=mask)

def grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

def gaussian_blur(image, kernel_size=5):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

def canny_edge_detection(image, low_threshold=50, high_threshold=150):
    return cv2.Canny(image, low_threshold, high_threshold)

def region_of_interest(image):
    height, width = image.shape[:2]
    polygons = np.array([
        [(0, height), (width, height), (width, height // 2), (0, height // 2)]
    ])
    mask = np.zeros_like(image)
    cv2.fillPoly(mask, polygons, 255)
    return cv2.bitwise_and(image, mask)

def hough_transform(image):
    return cv2.HoughLinesP(image, rho=1, theta=np.pi/180, threshold=20, minLineLength=20, maxLineGap=30)

def draw_lines(image, lines):
    # Draw the lines on a copy of the original image (to preserve the original)
    line_image = np.copy(image) * 0  # Creating a blank to draw lines on
    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(line_image, (x1, y1), (x2, y2), (0, 0, 255), 10)
    # Merge the line image with the original image
    combo = cv2.addWeighted(image, 0.8, line_image, 1, 1)
    return combo

# Load the image
image = cv2.imread('3.png')
# Display the original image
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()

# Apply color selection
color_selected_image = color_selection(image)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(color_selected_image, cv2.COLOR_BGR2RGB))
plt.title('Color Selected Image')
plt.axis('off')
plt.show()

# Convert to grayscale
gray_image = grayscale(color_selected_image)
plt.figure(figsize=(10,10))
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

# Apply Gaussian smoothing
blur_image = gaussian_blur(gray_image)
plt.figure(figsize=(10,10))
plt.imshow(blur_image, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.axis('off')
plt.show()

# Perform Canny edge detection
edges = canny_edge_detection(blur_image)
plt.figure(figsize=(10,10))
plt.imshow(edges, cmap='gray')
plt.title('Canny Edges')
plt.axis('off')
plt.show()

# Determine the region of interest
roi_image = region_of_interest(edges)
plt.figure(figsize=(10,10))
plt.imshow(roi_image, cmap='gray')
plt.title('Region of Interest')
plt.axis('off')
plt.show()

# Apply Hough transform to find the lanes within the ROI
lines = hough_transform(roi_image)
final_image = draw_lines(image, lines)

# Display the image
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB))
plt.title('Final Image with Lane Lines')
plt.axis('off')
plt.show()
```
